```python
def test(fromLine):
  URL = ('http://apis.data.go.kr/B553766/smt-transfer/'
      'transfer?'
      '&numOfRows=100'
      '&pageNo=1'
      '&fromLine='+fromLine+
      '&serviceKey=dphVQ2qUOT7iUQUVCDCfJk%2Fny0ozSMFz2pBAPRN7qTZUL%2FitfIfLcjiOXDSCSUg6GB0ISikz%2FdUaxmiNX%2F4Rew%3D%3D')
  
  response = requests.get(URL)
  r = response.json()
  items = r['data']

  departure_name = [item['departure_name'] for item in items]
  before_line = [item['before_line'] for item in items]
  alight_position = [item['alight_position'] for item in items]
  arrival_name = [item['arrival_name'] for item in items]
  after_line = [item['after_line'] for item in items]
  arrival_direction = [item['arrival_direction'] for item in items]
  boarding_position = [item['boarding_position'] for item in items]
  take_time = [item['take_time'] for item in items]

  departure_name = pd.Series(departure_name)
  before_line = pd.Series(before_line)   
  alight_position = pd.Series(alight_position)
  arrival_name = pd.Series(arrival_name)
  after_line = pd.Series(after_line)
  arrival_direction = pd.Series(arrival_direction)
  boarding_position = pd.Series(boarding_position)
  take_time = pd.Series(take_time)

  df = pd.DataFrame({ 'departure_name': departure_name, 'before_line': before_line, 'alight_position': alight_position, 'arrival_name': arrival_name, 'after_line': after_line, 'arrival_direction': arrival_direction, 'boarding_position': boarding_position, 'take_time': take_time})
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['1', '2', '3', '4', '5', '6', '7', '8', '9']), outputs=['dataframe'])
iface.launch(debug=True)
```
